```python
# An谩lisis del D贸lar Oficial Argentino
# Datos descargados desde la API del BCRA

import pandas as pd
import matplotlib.pyplot as plt

# Leer los datos desde el CSV limpio
df = pd.read_csv("data/dolar_oficial_limpio.csv")
df["fecha"] = pd.to_datetime(df["fecha"])

# Mostrar las primeras filas
df.head()
```

```python
#  Gr谩fico de la evoluci贸n del d贸lar oficial

plt.figure(figsize=(12, 6))
plt.plot(df["fecha"], df["valor_usd"], label="D贸lar Oficial", color="blue")
plt.title("Evoluci贸n del D贸lar Oficial (BCRA)")
plt.xlabel("Fecha")
plt.ylabel("Valor en ARS")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
